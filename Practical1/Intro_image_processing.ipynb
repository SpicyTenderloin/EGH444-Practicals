# 7 Data Classes & Image Types

```python
im = cv2.imread('baboon.jpg')
im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)  # CONVERT BGR TO GRAY
print(im_gray.dtype)

# Operations on the image in integer format vs in float format
im1 = (im_gray / 128).astype(np.uint8)*255
print(im.dtype)
im2 = (im_gray / 128).astype(np.float64)*255
print(im2.dtype)
plt.figure();plt.imshow(im1, cmap='gray');plt.axis('off');plt.title('uint8 Image')
plt.figure();plt.imshow(im2, cmap='gray');plt.axis('off');plt.title('float64 Image')
plt.figure();plt.imshow(im2 - im1, cmap='gray');plt.axis('off');plt.title('Difference casting uint8 or float64')

# Convert image to logical (binary) image
im_log = im_gray > 120
print(im_log.dtype)
plt.figure();plt.imshow(im_log, cmap='gray');plt.axis('off');plt.title('Logical Image (Binary)')

# Normalize intensities between 0 and 1
im_scaled = cv2.normalize(im_gray, None, 0, 1, cv2.NORM_MINMAX)
plt.figure();plt.imshow(im_scaled, cmap='gray');plt.title('Normalized Image (0-1 Range)');plt.axis('off')

# Low contrast image using thresholding [100, 200]
im_lowcon = np.clip(im_gray, 100, 200)
plt.figure();plt.imshow(im_lowcon, cmap='gray');plt.title('Low Contrast Image');plt.axis('off')
```
